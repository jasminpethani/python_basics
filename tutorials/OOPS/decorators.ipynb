## Decorator
- decorators starting with '@' keyword
- e.g. @abstractmehod
- above method used commonly.

```python
def main_welcome_method(func):
    def sub_welcome_method():
        print("Welcome to first line")
        func()
        print("thank you for msg!!")
    return sub_welcome_method()

# main_welcome_method()
```

```python
@main_welcome_method
def couse_on_udemy():
    print("This is very good course")
```

```python
def repeat(n):
    def decorator(func):
        def wrapper(*args, **kargs):
            for _ in range(n):
                func(*args, **kargs)
        return wrapper
    return decorator
```

```python
@repeat(3)
def say_hello():
    print("hello!!")
```
