## Generators use to loop through iterators, it uses 'yield' keyword to lazy loading iterator for memory mangement


```python
my_list=[1,2,3,4,5,6]

def square(n):
    for i in range(3):
        yield i**2
        
```

```python
square(3)
```

```python
for i in square(3):
    print(i)
```

```python
def print_somethig():
    yield 1
    yield 2
    yield 3

```

```python
print_somethig()

for i in print_somethig():
    print(i)
```

```python
# reading large file

def read_largefile(filepath): 
    with open(filepath, 'r') as file: 
        for line in file:
            yield line

try: 
    filepath = "large_file.txt"
    for line in read_largefile(filepath): 
        print(line.strip())
except FileNotFoundError:
    print("file doesn't exists just create it.")
    
```
