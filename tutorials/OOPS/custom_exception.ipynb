## Custom exception (Raise and throw an exception)

```python
class dobException(Exception): 
    pass
```

```python
try: 
    year=int(input("Enter your date of birth year: "))
    age=2026-year

    if age<=30 and age >= 20:
        print("you're teenager")
    else:
        raise dobException
except dobException:
    print("age is not <=30 or >=20 criteria.")
    
```
