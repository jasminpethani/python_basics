```python
from math import pi
from abc import ABC, abstractmethod

## abstrac class and method
class Shape(ABC): 
    @abstractmethod
    def area(self): pass
```

```python
# derived class 1
class Circle(Shape):
    def __init__(self, radius):
        self.radius=radius

    def area(self): 
        return pi * (self.radius ** 2)

# derived class 2
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width=width
        self.height=height
    
    def area(self):
        return self.width * self.height

shape=Rectangle(25, 30) 
shape.area()
```
