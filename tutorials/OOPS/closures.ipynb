## closures == nested functions

```python
def welcome_msg():
    msg="welcome"
    def sub_welcome_msg():
        print(f"{msg.upper()} to python world")
        print("another welcoming message")
    return sub_welcome_msg()
```

```python
welcome_msg()
```

```python
def another_example(name):
    print("initialize content")
    def sub_method():
        print(f"Welcome to the world of python!! {name.upper()}")
    return sub_method

another_example("jasmin")()
```

```python
def print_method(func):
    def welcome_method():
        print(f"Welcome to python learning session")
        func("how are you")
        print("THE END!!!")
    return welcome_method() 

```

```python
def here_try_me(str):
    print(f"inbetween printing..\n {str}")

print_method(here_try_me) 
```
