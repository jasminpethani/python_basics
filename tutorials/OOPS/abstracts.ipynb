```python
## ABC (Abstract base class )
from abc import ABC, abstractmethod

class Vehicle(ABC): 
    def run():
        print("running....")

    @abstractmethod
    def average(self): 
        return "this will give average"
```

```python
#Can't instantiate abstract class Vehicle without an implementation for abstract method 

# object=Vehicle() 
```

```python
class Car(Vehicle):
    def average(self):
        return 4
```

```python
tesla=Car()
print(tesla) 
print(tesla.average())
```
