## Functions in python


```python
def any_args(*args):
    print(f"{len(args)}")

any_args(1,2,3,4,5,10)
```

```python
def print_dict(**args):
    for key, value in args.items():
        print(f"key: {key}, value: {value}")


print_dict(name="jasmin", age=35, country="india")
```

```python
def word_count_frequency(filepath): 
    word_count={}
    with open(filepath, 'r') as file: 
        for line in file:
            words = line.split()
            for word in words:
                word = word.lower().strip('.,!@?;:"')
                word_count[word] = word_count.get(word, 0) + 1


    return word_count


word_count_frequency('sample.txt')

```
