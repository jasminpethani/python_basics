### Custom iterators 
iter() method to create iterator from list intilizing lazily.

```python
my_list=[1,2,3,4,5,6]
my_list
```

```python
iterator=iter(my_list)
type(iterator)
```

```python
try: 
   for i in next(iterator):
        print(f"element value: {i}")
except StopIteration: 
    print("last element is reached")
```
