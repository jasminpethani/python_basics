```python
try: 
    a=b 
except NameError as ex:
    print(ex)
```

```python
try:
    result=1/2
    a=b
except ZeroDivisionError as ex: 
    print(ex)
    print("devision")
except Exception as ex1: 
    print(ex1)

# 'Exception' class is parent of every exception

```

```python
# try, except, else block, finally

try: 
    num=int(input("enter number: "))
    result=10/num 
except ValueError:
    print("no a number")
except ZeroDivisionError as er:
    print("denominator should be > 0")
else: 
    print(f"result is {result}")
finally: 
    print("this is always execute with error or NOT")
```
