## List comprehension
-  basic  syntax [ expression for x in list  ]
- conditional syntax [ expression for x in list if condition ]

```python
## List Comprehence 

numbers = [] 
for x in range(10):
    numbers.append(x**2)

print(numbers)


# ------- - 

print(
    [x ** 2 for x in range(10)]
)

print(
    [num for num in range(10) if num % 2 == 0]
)
```

```python
words_list = ["hello", "world", "how", "are", "you", "?"]
word_lengths = [len(word) for word in words_list]
print(word_lengths)
```
