```python
student = {'name': 'jasmin', 'age': '35', 'is_employed': True}

print(student)
```

```python
student_copy = student 
student_copy['name'] = "Jasmin"

print(student)
print(student_copy)
```

```python
student_copy1 = student.copy()  #shalow copy

student_copy1['name'] = "JASMIN"
print(student_copy1)
print(student)
```

```python
# Iteration
 
for key,value in student.items():
    print(f"{key} is {value}")
```

```python
## comprehensive dictionaries
squares={x:x**3 for x in range(5)}
print(squares)
```

```python
numbers=[1,2,2,1,3,3,4,4,4,4,5,6, 1,6,6,7,7]
print(numbers)

result = {}

for num in numbers: 
    if num in result:
        result[num] += 1
    else:
        result[num] = 1
print(result)
```

```python
## Mearge dictionaries
dict1 = {'a': 2, 'b': 5}
dict2 = {'b': '10', 'c': 1}

merged_dict = {**dict1, **dict2}
print(merged_dict)
```
