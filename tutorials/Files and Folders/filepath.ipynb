### File path example


```python
import os
print(os.getcwd())
```

```python
newdir="new_dir"
os.mkdir(newdir) # new directory created 
```

```python
## listing files and directory
items=os.listdir('.')
print(items)

for item in items:
    fullpath=os.path.join(os.getcwd(), item)
    print(os.path.isfile(fullpath))

print(os.path.join(os.getcwd(), 'dir01', 'filename.txt'))
```

```python
filepath = 'example.txt'

if os.path.isfile(filepath):
    print(f"{filepath} is file")
elif os.path.isdir(filepath):
 print(f"{filepath} is directory")
else:
    print(f"{filepath} is neither directory nor file")
```
