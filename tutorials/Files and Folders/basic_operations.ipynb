### File & Directory operations
- which helps in RAG application creation to train model to read org. level files/folders
- appending, reading, writing modes for file


```python
with open('example.txt', 'r') as file:
    content = file.read()
    print(content)
```

```python
with open('example.txt', 'r') as file:
   for line in file: 
    print(line.strip()) # strip() removes extra new line characters
```

## Writing something into file



```python
with open("example.txt", 'w') as file: 
    file.write('hello world\n')
    file.write('this is new line again') 
```

```python
with open("example.txt", 'a') as file: 
    file.write('\nappending new line again.... testing')
     
```

```python
## read the lines, words and characters 
with open("example.txt", 'r') as file: 
    lines = 0 
    nwords = 0
    characters = 0
    for line in file: 
        lines += 1
        words=line.split(' ') 
        for word in words:
            nwords += 1
            characters += len(word)

    print(f"Number of lines {lines}, words {nwords}, and characters {characters}") 
        
```
